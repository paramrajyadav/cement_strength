```python
def ridge_regression():
    from sklearn.linear_model import Ridge
    from sklearn.model_selection import GridSearchCV

    # Define the model
    model = Ridge(random_state=42)

    # Define the grid search parameters
    param_grid = {
        'alpha': [0.1, 1, 10],
        'max_iter': [1000, 5000, 10000],
        'tol': [0.001, 0.01, 0.1],
        'normalize': [True, False],
        'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
    }

    # Define the GridSearchCV object
    grid_search = GridSearchCV(model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

    # Fit the model with the GridSearchCV object
    grid_search.fit(X_train, y_train)

    # Print the best parameters and score
    print("Best Parameters:", grid_search.best_params_)
    print("Best Score:", grid_search.best_score_)

    # Fit the Ridge Regression with the best parameters
    ridge_reg = Ridge(random_state=42, **grid_search.best_params_)
    ridge_reg.fit(X_train, y_train)
    predict_clf = ridge_reg.predict(X_test)
    global ridge_error
    ridge_error = r2_score(y_test, predict_clf)
    return ridge_reg
```

```python
def linear_reg():
    from sklearn.model_selection import GridSearchCV
    from sklearn.linear_model import LinearRegression

    # Define the model
    model = LinearRegression()

    # Define the grid search parameters
    param_grid = {'fit_intercept': [True, False],
    'copy_X': [True, False],
    'n_jobs': [None, -1]}

    # Define the GridSearchCV object
    grid_search = GridSearchCV(model, param_grid=param_grid,verbose=2)

    # Fit the model with the GridSearchCV object
    grid_search.fit(X_train, y_train)
    
    best_model = LinearRegression(fit_intercept=grid_search.best_params_['fit_intercept'],
                            
                              copy_X=grid_search.best_params_['copy_X'],
                              n_jobs=grid_search.best_params_['n_jobs'])
    best_model.fit(X_train, y_train)
    predict_clf=best_model.predict(X_test)
    global linear_reg_error
    linear_reg_error = r2_score(y_test,predict_clf)
    return best_model
```
